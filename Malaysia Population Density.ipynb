```python
import folium
import geopandas as gpd
```

```python
gpkg_file = r"C:\Users\PC\Downloads\kontur_population_MY_20220630.gpkg\kontur_population_MY_20220630.gpkg"
density_gdf = gpd.read_file(gpkg_file, layer='population')
```

```python
json = r"C:\Users\PC\Downloads\geoBoundaries-MYS-ADM1_simplified.geojson"
json_gpd = gpd.read_file(json)
```

```python
json_gpd
```

```python
density_gdf
```

```python
json_gpd = json_gpd.to_crs(density_gdf.crs)
```

```python
pop_join = gpd.sjoin(density_gdf,json_gpd,predicate='intersects',how="inner")
```

```python
pop_join
```

```python
def style_function(feature):
    return {'fillColor': '#ffffff', 'color': '#ffffff', 'fillOpacity': 0.0, 'stroke': False}
```

```python
# Create a map centered on Kuala Lumpur
map = folium.Map(location=[3.1390, 101.6869], zoom_start=11, width=1000, height=1000)

# Add the intersect_gdf to the map as a GeoJSON layer
folium.GeoJson(pop_join).add_to(map)

# Display the map
map
```

```python
folium.Choropleth(
    geo_data=pop_join,
    data=pop_join,
    columns=['h3', 'population'],
    key_on='feature.properties.h3',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.1,
    legend_name='Population Density',
    highlight=True,
    style_function=style_function,
    tooltip = ['Population Density: ','population']
).add_to(map)

# Display the map
map
```

```python
pop_join.to_csv("population_in_malaysia.csv", index=False)
```
